### import model

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import numpy as np
```

```python
fname = 'finalmodel'
model = Doc2Vec.load(fname)
```

### Similarity Query

- doc to doc : docvecs.mostsimilar
- word to word : wv.mostsimilar

```python
model.docvecs.most_similar('Lim, Chiehyeon(임치현)',topn=50)
```

```python
model.wv.most_similar('service',topn=30)
```

### documents vector, docvecs

- model.docvecs.doctags : documents tag의 정보를 알려준다
- index2entity : 전체 documents tag를 조회시켜준다.
- index_to_doctag : index를 넣으면 해당하는 documents를 조회시켜준다
- model.docvecs[0] : 해당 index를 가지는 doctag의 vector값을 조회시켜준다
- vectors_docs : 전체 documents 에 대해서 백터값을 조회시켜준다

```python
model.docvecs.doctags
```

```python
model.docvecs.index2entity[0:5]
```

```python
model.docvecs.index_to_doctag(0)
```

```python
model.docvecs[0]
```

```python
print(model.docvecs.vectors_docs.shape)
model.docvecs.vectors_docs
```

### Wordvec tutorial

- vocab : 사전형태의 단어장을 보여준다
- index2entity : 전체 단어들의 list를 조회시켜준다
- word_vec : 단어에 대한 vector값을 조회시켜준다

```python
model.wv.vocab
```

```python
model.wv.index2word[0]
```

```python
model.wv.index2entity[0:5]
```

```python
model.wv.word_vec('using')
```

```python
model.wv.vectors
```

### Wordvector 와 docvec를 합치는 방법

```python
docvec = model.docvecs.vectors_docs
wordvec = model.wv.vectors
totalvector = np.vstack((docvec,wordvec))
print(totalvector.shape)
```
