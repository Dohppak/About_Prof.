```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
import json
```

```python
# mongo db클라이언트와, db를 선언한다. 본 프로젝트의 db명을 aboutdepart라고 할때
client = MongoClient()
db = client.aboutdepart
```

```python
prof_col = ['Name','id','Lab','Name','ResearchInterest','Website','department','imglink']
paper_col = ['Title','Cited by','Link', 'Source title','Year','Authors']
```

```python
for i in list(db.finaldb.find()):
    document ={}
    paper = []
    
    for j in prof_col:
        document[j]=i[j]
        
    a= i['Abstract']
    b= i['Cited by']
    c= i['Link']
    d= i['Source title']
    e= i['Year']
    f= i['Authors']
    h= i['Title']
    
    for q,w,e,r,t,y,u in zip(a,b,c,d,e,f,h):
        journal = {}
        journal['Title'] = u
        journal['Cited by'] =w
        journal['Link'] =e
        journal['Source title'] = r
        journal['Year'] = t
        journal['Authors'] = y
        paper.append(journal)
    
    document['paper']=paper
    db.profinfo.insert_one(document)
```

```python
def time_series(df):
    return (df[['Title','Year']].groupby(['Year']).count().to_dict(),
    df[['Cited by','Year']].replace('nan', 0).groupby(['Year']).sum().fillna(0).to_dict())
```

```python
time_series(df_paper)
```

```python
from collections import Counter

def most_freq_coauthor(df, n):
    authors = []
    for i in df['paper']:
        j = i['Authors'].split(", ")
        authors = authors + j

    author_counts = Counter(authors)
    
    return author_counts.most_common(n)
```

```python
def most_freq_source(df, n):

    source = []
    for i in df['paper']:
        source.append(i['Source title'])

    source_counts = Counter(source)
    
    return source_counts.most_common(n)
```

```python
most_freq_coauthor(test, 5)
```
