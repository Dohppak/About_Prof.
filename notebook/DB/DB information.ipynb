````python
prof_col = ['Name','id','Lab','Name','ResearchInterest','Website','department','imglink']
paper_col = ['Title','Cited by','Link', 'Source title','Year','Authors']
````

````python
for i in list(db.finaldb.find()):
    document ={}
    paper = []
    
    for j in prof_col:
        document[j]=i[j]
        
    a= i['Abstract']
    b= i['Cited by']
    c= i['Link']
    d= i['Source title']
    e= i['Year']
    f= i['Authors']
    h= i['Title']
    
    for q,w,e,r,t,y,u in zip(a,b,c,d,e,f,h):
        journal = {}
        journal['Title'] = u
        journal['Cited by'] =w
        journal['Link'] =e
        journal['Source title'] = r
        journal['Year'] = t
        journal['Authors'] = y
        paper.append(journal)
    
    document['paper']=paper
    db.profinfo.insert_one(document)
````

````python
def cNumber(inp):
    df = pd.DataFrame(inp)
    df1 = df[['Title','Year']].groupby(['Year']).count()
    df2 = df[['Cited by','Year']].replace('nan', 0).groupby(['Year']).sum().fillna(0)
    df3 = pd.merge(df1, df2)
    return df3
````

````python
from collections import Counter

inp = test['paper']
authors = []
for i in inp:
    j = i['Authors'].split(", ")
    authors = authors + j

author_counts = Counter(authors)
temp = []
for k in author_counts.most_common(5):
    documents = {}
    documents['coName'] = k[0]
    documents['cNumber'] = k[1]
    temp.append(documents)
````
